# DEGISKEN STANDARDIZASYONU (VERI STANDARDIZASYONU)

```python
import numpy as np
import pandas as pd
V1 = np.array([1,3,6,5,7])
V2 = np.array([7,7,5,8,12])
V3 = np.array([6,12,5,6,14])
df = pd.DataFrame({"V1": V1,
                  "V2": V2,
                  "V3": V3})
df = df.astype(float)
df
```

### Standardizasyon

```python
from sklearn import preprocessing
```

```python
preprocessing.scale(df) #scale donusturme yapar. hepsine standart donusum uygular.
```

### Normalizasyon

```python
preprocessing.normalize(df) #0 ve 1 arası donusturme ypaar.
```

### Min-Max donusumu

```python
scaler = preprocessing.MinMaxScaler(feature_range = (10,20)) #10-20 arasi bir donusturucu olusturduk.
```

```python
scaler.fit_transform(df) #donusturucuyu fit transform ile uyguladık.
```

# DEGISKEN DONUSUMLERI

```python
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

```python
#Her bir standardizasyon bir degisken donusumudur.
```

### 0-1 Donusumu

```python
from sklearn.preprocessing import LabelEncoder

lbe = LabelEncoder()
```

```python
lbe.fit_transform(df["sex"])
```

```python
df["yeni_sex"] = lbe.fit_transform(df["sex"])
```

```python
df
```

### 1 ve digerleri(0) Dönüşümü

```python
df.head()
```

```python
import numpy as np
df["yeni_day"] = np.where(df["day"].str.contains("Sun"),1,0)
```

```python
df
```

### Cok sinifli donusum

```python
from sklearn.preprocessing import LabelEncoder

lbe = LabelEncoder()
```

```python
lbe.fit_transform(df["day"])
```
